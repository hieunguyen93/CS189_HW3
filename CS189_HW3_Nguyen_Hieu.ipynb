```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(26)

size = 100
mu_x = 3
var_x = 9
mu_y = 4
var_y = 4

x_sample = np.random.normal(mu_x, var_x, size)
y_sample = np.random.normal(mu_y, var_y, size)

sample_points = np.array([np.array((x, 0.5 * x + y)) for (x, y) in zip(x_sample, y_sample)])
```

### (a): compute the mean of the sample

```python
# (a): compute the mean of the sample

sample_mean = np.mean(sample_points, axis=0)

print('Mean of the Sample:')

print(sample_mean)
```

### (b): compute the 2x2 covariance matrix of the sample

```python
# (b): compute the 2x2 covariance matrix of the sample

sample_covariance = np.cov(sample_points.T)

print('2x2 Covariance Matrix of the Sample')

print(sample_covariance)
```

### (c) compute the eigenvectors and eigenvalues of the covariance matrix

```python
# (c) compute the eigenvectors and eigenvalues of the covariance matrix

eigenvalues, eigenvectors = np.linalg.eig(sample_covariance)

print('Eigenvalues of the Covariance Matrix:')

print(eigenvalues)

print("-------------------")

print('Eigenvectors of the Covariance Matrix:')

print(eigenvectors)
```

```python
np.random.seed (26)

def split_train_val_data(data, labels, val_size):
    num_items = len(data)
    assert num_items == len(labels)
    assert val_size >= 0
    if val_size < 1.0:
        val_size = int(num_items * val_size)
    train_size = num_items - val_size
    idx = np.random.permutation(num_items)
    data_train = data[idx][:train_size]
    label_train = labels[idx][:train_size]
    data_val = data[idx][train_size:]
    label_val = labels[idx][train_size:]
    return data_train, data_val, label_train, label_val

class LDA:
    def __init__(self):
        self.mu = None
        self.sigma = None
        self.Pi = None
    
    def fit(self, X, y):
        
        N, labels = len(y), np.unique(y)
        d = X.shape[1]
        C = len(labels)
        mu = np.zeros((C, d))
        Pi = np.zeros(C)
        sigma = np.zeros((d, d))
        for i, label in enumerate(labels):
            X_i = []
            for index, result in enumerate(y):
                if result == label:
                    X_i.append(X[index])
            X_i = np.array(X_i)
            #print(len(X_i))
            Pi[i] = len(X_i) / N
            mu[i, :] = np.mean(X_i, axis=0)
            sigma += np.cov(X_i.T)*X_i.shape[0]
        sigma /= N
        self.mu = mu
        self.sigma = sigma
        self.Pi = Pi
    
    def predict(self, X):
        
        d = len(self.sigma)
        self.sigma += 1e-15*np.array(np.eye(d))
        
        L1 = self.mu.dot(np.linalg.solve(self.sigma, X.T)).T
        
        L2 = 1/2*np.diag(self.mu.dot\
                         (np.linalg.solve(self.sigma, self.mu.T)))
        
        L = L1 - L2 - np.log(self.Pi)
        
        pred = np.argmax(L.T, axis=0)
        

        return pred
    
    def accuracy(self, X, y):
        N = len(y)
        correct = 0
        pred = self.predict(X)
        for i in range(N):
            if pred[i] == y[i]:
                correct += 1
        return correct/N, pred
    
mnist_train_X, mnist_val_X, mnist_train_y, mnist_val_y = split_train_val_data(cn_mnist_X, mnist_y, 10000)


training_size = np.array([100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000])
lda_accuracy = np.zeros(len(training_size))

for i, N in enumerate(training_size):
    lda = LDA()
    lda.fit(mnist_train_X[:N,:], mnist_train_y[:N])
    lda_accuracy[i], _ = lda.accuracy(mnist_val_X, mnist_val_y)

lda_accuracy

```

```python
class QDA:
    def __init__(self, a):
        self.mu = None
        self.sigma = None
        self.Pi = None
        self.a = a
    
    def fit(self, X, y):
        N, labels = len(y), np.unique(y)
        #print(labels)
        d = X.shape[1]
        C = len(labels)
        sigma = [np.zeros([d, d]) for i in range(N)]
        mu = np.zeros([C, d])
        Pi = np.zeros([C])
        for i, label in enumerate(labels):
            # Get each classes' statistics
            X_i = []
            for index, result in enumerate(y):
                if result == label:
                    X_i.append(X[index])
            X_i = np.array(X_i)
            Pi[i] = len(X_i) / N
            mu[i, :] = np.mean(X_i, axis=0)
            sigma[i] = np.cov(X_i.T) + self.a*np.eye(d)
        self.mu = mu
        self.sigma = sigma
        self.Pi = Pi
        
            
    def predict(self, X):
        C = len(self.Pi)
        L = np.zeros((C, len(X)))
        mu = self.mu
        sigma = self.sigma
        Pi = self.Pi
        for i in range(C):
            #print('start')
            L[i,:] = multivariate_normal.logpdf(X, mean=mu[i], cov=sigma[i])
            L[i,:] += np.log(Pi[i])
        pred = np.argmax(L, axis=0)
        return pred
    
    def accuracy(self, X, y):
        N = len(y)
        correct = 0
        pred = self.predict(X)
        for i in range(N):
            if pred[i] == y[i]:
                correct += 1
        #print('accuracy', correct/N)
        return correct/N, pred
    
training_size = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]

qda_accuracy = np.zeros(len(training_size))


for i, N in enumerate(training_size):
    qda = QDA(1e-8)
    qda.fit(mnist_train_X[:N,:], mnist_train_y[:N])
    qda_accuracy[i], _ = qda.accuracy(mnist_val_X, mnist_val_y)

qda_accuracy
```

```python
spam = io.loadmat("data/spam_data.mat")
spam_X, spam_y = spam['training_data'].astype(float), spam['training_labels']
spam_test_x = spam['test_data'].astype(float)


spam_train_x, spam_val_x, spam_train_y, spam_val_y = split_train_val_data(spam_X, spam_y, 500)



# Computes prior, mean, and covariance
def spam_train_lda(train_x, train_y):
    n,d = train_x.shape
    priors = np.zeros((2,1))
    covariance = np.zeros((d,d)).astype(np.float32)
    mean_samples = np.zeros((2, d))
    train_y = train_y.reshape(-1,)
    
    for spam in [0,1]:
        ## Compute priors
        prob = (train_y == spam).astype(np.int32).sum() / n
        priors[int(spam)] = prob

        data = train_x[train_y == spam, :]        
        mean_samples[int(spam),:] = data.mean(axis=0)
        covariance += np.cov(data.T)
        
    covariance /= 2
    return priors, mean_samples, covariance

# Train
lda_priors, lda_means, lda_covariance = spam_train_lda(spam_train_x, spam_train_y)

N_train = spam_train_x.shape[0]
out_train = np.zeros((N_train, 2))

N_val = spam_val_x.shape[0]
out_val = np.zeros((N_val, 2))
        
for class_y in [0,1]:
    prior = lda_priors[class_y]
    mean = lda_means[class_y, :].reshape(-1, 1)

    w = np.linalg.solve(lda_covariance.T, mean)
    alpha = -0.5 * w.T.dot(mean) + np.log(prior)

    out_train[:, class_y] = (spam_train_x.dot(w) + alpha).reshape(-1,)
    out_val[:, class_y] = (spam_val_x.dot(w) + alpha).reshape(-1,)

train_pred = np.argmax(out_train, axis=1).reshape(-1, 1)
val_pred = np.argmax(out_val, axis=1).reshape(-1, 1)

def computeDiagonals(X, X_T):
    N,D = X.shape
    out_diag = np.zeros((N,1))
    
    for n in range(N):
        i = X[n, :].reshape(1, -1)
        i_t = X_T[:, n].reshape(-1, 1)
        out_diag[n] = i.dot(i_t)
    
    return out_diag

def spam_train_qda(train_x, train_y):
    n,d = train_x.shape
    priors = np.zeros((2,1))
    covariances = np.zeros((2,d,d)).astype(np.float32)
    mean_samples = np.zeros((2, d))
    train_y = train_y.reshape(-1,)
    
    for spam in [0,1]:
        ## Compute priors
        prob = (train_y == spam).astype(np.int32).sum() / n
        priors[int(spam)] = prob

        data = train_x[train_y == spam, :] 
        
        mean_samples[int(spam),:] = data.mean(axis=0)
        covariances[int(spam), :, :] = np.cov(data.T)
        
    return priors, mean_samples, covariances

# Train
qda_priors, qda_means, qda_covariances = spam_train_qda(spam_train_x, spam_train_y)

N_train = spam_train_x.shape[0]
out_train = np.zeros((N_train, 2))

N_val = spam_val_x.shape[0]
out_val = np.zeros((N_val, 2))

for spam in [0,1]:
    prior = qda_priors[spam]
    mean = qda_means[spam, :]
    covariance = qda_covariances[spam, :, :]

    u, sig, v = np.linalg.svd(covariance)
    covariance = covariance + (sig.min()*(10**-1)) * np.eye(*covariance.shape)
    
    alpha = -0.5 * np.linalg.det(covariance) + prior

    mean_centered_train = spam_train_x - mean
    weight_train = np.linalg.solve(covariance, mean_centered_train.T)
    prediction_train = -0.5*computeDiagonals(mean_centered_train, weight_train) + alpha
    out_train[:, spam] = prediction_train.reshape(N_train,)

    mean_centered_val = spam_val_x - mean
    weight_val = np.linalg.solve(covariance, mean_centered_val.T)
    prediction_val = -0.5*computeDiagonals(mean_centered_val, weight_val) + alpha
    out_val[:, spam] = prediction_val.reshape(N_val,)


train_pred = np.argmax(out_train, axis=1).reshape(-1, 1)
val_pred = np.argmax(out_val, axis=1).reshape(-1, 1)


from save_csv import results_to_csv
N_test = spam_test_x.shape[0]
out_test = np.zeros((N_test, 2))
    
for class_y in [0,1]:
    prior = lda_priors[class_y]
    mean = lda_means[class_y, :].reshape(-1, 1)

    w = np.linalg.solve(lda_covariance.T, mean)
    alpha = -0.5 * w.T.dot(mean) + np.log(prior)

    out_test[:, class_y] = (spam_test_x.dot(w) + alpha).reshape(-1,)

test_pred = np.argmax(out_test, axis=1).reshape(-1,)
results_to_csv(test_pred)
```
